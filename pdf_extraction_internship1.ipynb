```python
import pdfplumber
import re
from transformers import pipeline
```

```python
def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Load the PDF file
pdf_path = '/content/sample/c669abb4-f485-4880-8973-cc7fdfeee22e.pdf'
pdf_text = extract_text_from_pdf(pdf_path)
print(pdf_text)
```

```python
# Load a pre-trained model from Hugging Face
nlp_pipeline = pipeline('text-generation', model='distilgpt2')
```

```python
# Use the pipeline to analyze the PDF text
def analyze_text_with_llm(text):
    generated_text = nlp_pipeline(text, max_length=500, do_sample=True, top_p=0.95, top_k=50,truncation=True)
    return generated_text[0]['generated_text']

analysis = analyze_text_with_llm(pdf_text)
print(analysis)
```

```python
import re

def extract_details_from_analysis(analysis):
    details = {}

    # Extracting details following "Bill To:"
    bill_to_section = re.search(r'Bill To:(.*?)(?=\n\n|$)', analysis, re.DOTALL)
    if bill_to_section:
        bill_to_details = bill_to_section.group(1).strip()
        # Extract specific fields from the "Bill To:" section, handling potential missing matches
        details['Invoice Date'] = re.search(r'Invoice Date:\s*(.*)', bill_to_details)
        details['Invoice Date'] = details['Invoice Date'].group(1).strip() if details['Invoice Date'] else None

        details['Place of Supply'] = re.search(r'Place of Supply:\s*(.*)', bill_to_details)
        details['Place of Supply'] = details['Place of Supply'].group(1).strip() if details['Place of Supply'] else None

        details['Enquire id'] = re.search(r'Enquire id:\s*(.*)', bill_to_details)
        details['Enquire id'] = details['Enquire id'].group(1).strip() if details['Enquire id'] else None

        details['Ship To'] = re.search(r'Ship To:\s*(.*)', bill_to_details)
        details['Ship To'] = details['Ship To'].group(1).strip() if details['Ship To'] else None

        details['Ph'] = re.search(r'Ph:\s*(.*)', bill_to_details)
        details['Ph'] = details['Ph'].group(1).strip() if details['Ph'] else None

        details['Email'] = re.search(r'(\S+@\S+)', bill_to_details)
        details['Email'] = details['Email'].group(1).strip() if details['Email'] else None

    # Extracting Item and Amount Payable
    details['Item'] = re.search(r'Item\s+([\s\S]*?)(?=\n|\Z)', analysis)
    details['Item'] = details['Item'].group(1).strip() if details['Item'] else None

    details['Amount Payable'] = re.search(r'Amount Payable:\s*(.*)', analysis)
    details['Amount Payable'] = details['Amount Payable'].group(1).strip() if details['Amount Payable'] else None

    return details

# Example usage
 # Replace with the actual text extracted from the PDF
details = extract_details_from_analysis(analysis)
print("Customer Details:")
print("Invoice Date:", details['Invoice Date'])
print("Place of Supply:", details['Place of Supply'])
print("Enquire id:", details['Enquire id'])
print("Ship To:", details['Ship To'])
print("Ph:", details['Ph'])
print("Email:", details['Email'])
print("\nProduct Name:")
print("Item:", details['Item'])
print("\nTotal Amount:")
print("Amount Payable:", details['Amount Payable'])
```
