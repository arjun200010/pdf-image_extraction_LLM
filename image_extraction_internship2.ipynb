```python
import easyocr
from PIL import Image
import re
from google.colab import files
```

```python
uploaded = files.upload()
image_path = next(iter(uploaded))
```

```python
def extract_text_from_image(image_path):
    reader = easyocr.Reader(['en'])  # Initialize EasyOCR reader
    result = reader.readtext(image_path)  # Perform OCR on the image
    text = ' '.join([item[1] for item in result])  # Combine text from all detected areas
    return text
extracted_text = extract_text_from_image(image_path)
print(extracted_text)
```

```python
def extract_details_from_text(text):
    details = {}

    # Extracting details following "Bill To:"
    bill_to_section = re.search(r'BIII To:(.*?)(?=\n\n|$)', text, re.DOTALL)
    if bill_to_section:
        bill_to_details = bill_to_section.group(1).strip()
        # Extract specific fields from the "Bill To:" section, handling potential missing matches
        details['Invoice Date'] = re.search(r'Involce Date:\s*(.*)', bill_to_details) # Fixed typo here: 'Involce Date' -> 'Invoice Date'
        details['Invoice Date'] = details['Invoice Date'].group(1).strip() if details['Invoice Date'] else None

        details['Place of Supply'] = re.search(r'Place of Supply:\s*(.*)', bill_to_details)
        details['Place of Supply'] = details['Place of Supply'].group(1).strip() if details['Place of Supply'] else None

        details['Ship To'] = re.search(r'Shlp To:\s*(.*)', bill_to_details)
        details['Ship To'] = details['Ship To'].group(1).strip() if details['Ship To'] else None

        details['Ph'] = re.search(r'Ph:\s*(.*)', bill_to_details)
        details['Ph'] = details['Ph'].group(1).strip() if details['Ph'] else None

        details['Email'] = re.search(r'(\S+@\S+)', bill_to_details)
        details['Email'] = details['Email'].group(1).strip() if details['Email'] else None

    # Extracting Item and Amount Payable
    details['Item'] = re.search(r'Item\s+([\s\S]*?)(?=\n|\Z)', text)
    details['Item'] = details['Item'].group(1).strip() if details['Item'] else None

    details['Amount Payable'] = re.search(r'Amount Payable:\s*(.*)', text)
    details['Amount Payable'] = details['Amount Payable'].group(1).strip() if details['Amount Payable'] else None

    return details

# Extract text from the uploaded image
text = extract_text_from_image(image_path)
details = extract_details_from_text(text)

print("Customer Details:")
print("Invoice Date:", details['Invoice Date']) # Accessing with the corrected key
print("\nPlace of Supply:", details['Place of Supply'])
print("\nShip To:", details['Ship To'])
print("\nPh:", details['Ph'])
print("\nEmail:", details['Email'])
print("\nProduct Name:")
print("Item:", details['Item'])
print("\nTotal Amount:")
print("Amount Payable:", details['Amount Payable'])
```
